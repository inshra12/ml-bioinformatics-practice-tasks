```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/cgpa dataset.csv');
data.head(3)
```

```python
x= data.iloc[:,:-1]
y=data['placed']
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='cgpa',y='score',data=data,hue='placed')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=6)
```

```python
from sklearn.linear_model import Perceptron
model=Perceptron()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)*100,model.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(5,4))
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=model)
plt.show()
```
